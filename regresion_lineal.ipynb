```python
# upload the data csv file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
df = pd.read_csv('regresion_lineal_data.csv')
df.head()
```

```python
# Crea un modelo de regresión lineal
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Define las variables independientes (características) y la variable dependiente (objetivo)
X = df[['metro']]  # Variable independiente (característica)
y = df['precio']   # Variable dependiente (objetivo)

# Entrena el modelo con los datos
model.fit(X, y)

# Realiza predicciones
y_pred = model.predict(X)

# Visualiza los resultados
plt.scatter(X, y, label='Datos reales')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicciones')
plt.xlabel('Variable independiente metros')
plt.ylabel('Variable dependiente precio')
plt.legend()
plt.show()

# Coeficiente de la regresión y término independiente
print('Coeficiente:', model.coef_)
print('Término independiente:', model.intercept_)
```
